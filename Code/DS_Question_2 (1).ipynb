# **Full code with all visualizations**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file once
df_full = pd.read_csv("labeled_financial_news_2M_clean.csv")


# Overall Financial News Analysis (2011-2020)


# Filter for years between 2011 and 2020 using the 'year' column
df_year_filter = df_full[(df_full['year'] >= 2011) & (df_full['year'] <= 2020)]

# Loop through each year and count sentiment labels
for year in range(2011, 2021):
    yearly_data = df_year_filter[df_year_filter['year'] == year]
    sentiment_counts_yearly = yearly_data['sentiment'].value_counts()

    # Get counts for each sentiment or default to 0 if not available
    bullish = sentiment_counts_yearly.get('bullish', 0)
    bearish = sentiment_counts_yearly.get('bearish', 0)
    neutral = sentiment_counts_yearly.get('neutral', 0)

    print(f"Year {year}:")
    print(f"  Bullish: {bullish}")
    print(f"  Bearish: {bearish}")
    print(f"  Neutral: {neutral}")
    print("-" * 30)

# Count overall sentiment totals from 2011 to 2020
sentiment_totals = df_year_filter['sentiment'].value_counts()
bullish_total = sentiment_totals.get('bullish', 0)
bearish_total = sentiment_totals.get('bearish', 0)
neutral_total = sentiment_totals.get('neutral', 0)

print("Total Headlines from 2011 to 2020:")
print(f"  Bullish: {bullish_total}")
print(f"  Bearish: {bearish_total}")
print(f"  Neutral: {neutral_total}")

# (Optional) Visualization using pre-calculated numbers (update these if needed)
sentiment_data = {'neutral': 839955, 'bullish': 392002, 'bearish': 316847}
total_count = sum(sentiment_data.values())
df_vis = pd.DataFrame.from_dict(sentiment_data, orient='index', columns=['Count'])
df_vis['Percentage'] = (df_vis['Count'] / total_count * 100).round(2)
df_vis = df_vis.loc[['bullish', 'bearish', 'neutral']]

print("\nSentiment Distribution Table (2011-2020):\n")
print(df_vis)

# Bar Chart for overall sentiment distribution
plt.figure(figsize=(8, 5))
plt.bar(df_vis.index, df_vis['Count'], color=['green', 'red', 'gray'])
plt.title("Number of Stock News Headlines by Sentiment (2011-2020)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# Pie Chart for overall sentiment distribution
plt.figure(figsize=(6, 6))
plt.pie(df_vis['Percentage'], labels=df_vis.index, autopct='%1.1f%%',
        startangle=140, colors=['green', 'red', 'gray'])
plt.title("Stock News Sentiment Distribution (2011-2020)")
plt.axis('equal')
plt.show()


# COVID-Specific Analysis (Using 'date' column for 2020)



df_covid = df_full.copy()


df_covid['date'] = pd.to_datetime(df_covid['date'], errors='coerce')


df_covid = df_covid.dropna(subset=['date'])


df_covid_2020 = df_covid[df_covid['date'].dt.year == 2020]


sentiment_counts_covid = df_covid_2020['sentiment'].value_counts()

print("\nSentiment Counts for 2020 (COVID):")
print(sentiment_counts_covid)


# COVID 2020 Visualizations


# Standardize the sentiment ordering and fill any missing values with 0.
covid_sentiments = ['bullish', 'bearish', 'neutral']
df_covid_vis = sentiment_counts_covid.reindex(covid_sentiments, fill_value=0).to_frame(name='Count')


total_covid = df_covid_vis["Count"].sum()
df_covid_vis["Percentage"] = (df_covid_vis["Count"] / total_covid * 100).round(2)

print("\nCOVID 2020 Sentiment Distribution Table:\n")
print(df_covid_vis)


plt.figure(figsize=(8, 5))
plt.bar(df_covid_vis.index, df_covid_vis['Count'], color=['green', 'red', 'gray'])
plt.title("COVID 2020: Stock News Headlines by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()


plt.figure(figsize=(6, 6))
plt.pie(df_covid_vis['Percentage'], labels=df_covid_vis.index, autopct='%1.1f%%',
        startangle=140, colors=['green', 'red', 'gray'])
plt.title("COVID 2020: Stock News Sentiment Distribution")
plt.axis('equal')
plt.show()
```
